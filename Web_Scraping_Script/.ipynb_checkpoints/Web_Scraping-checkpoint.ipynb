```python
import requests, bs4
import re
import json
import asyncio
import jmespath
from httpx import AsyncClient, Response
from parsel import Selector
from typing import List, Dict
import random
```

```python
cookies = {}
headers = {
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7",
        "Accept-Encoding": "gzip, deflate, br, zstd",
        "Accept-Language": "en-US,en;q=0.9,en-IN;q=0.8",
        "Cache-Control": "max-age=0",
        "Sec-Ch-Ua": '"Microsoft Edge";v="125", "Chromium";v="125", "Not.A/Brand";v="24"',
        "Sec-Ch-Ua-Mobile": "?0",
        "Sec-Ch-Ua-Platform": '"Windows"',
        "Sec-Fetch-Dest": "document",
        "Sec-Fetch-Mode": "navigate",
        "Sec-Fetch-Site": "same-origin",
        "Sec-Fetch-User": "?1",
        "Upgrade-Insecure-Requests": "1",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36 Edg/125.0.0.0"
    }
```

```python
def fetch_and_parse(url, headers, PrevCookies):
    try: 
        print(url)
        response = requests.get(url, headers=headers, cookies=PrevCookies)
        
        response_headers = response.headers
        response_cookies = response.cookies
        for cookie in response_cookies:
            cookies[cookie.name] = cookie.value
            
        if response.status_code == 200:
            soup = bs4.BeautifulSoup(response.text, 'html.parser')
            return soup
        else:
            print("Failed to fetch URL:", url)
            return None
    except requests.exceptions.RequestException as e:
        print("Error fetching URL:", e)
        return None
```

```python
def set_cookies(cookie_string,prevcookies):
    cookie_pairs = cookie_string.split('; ')
    for pair in cookie_pairs:
        key, value = pair.split('=', 1)
        prevcookies[key] = value
    cookies=prevcookies
```

```python
properties = []
href_list= []

nextURL = "https://www.realestate.com.au/buy/in-epping,+nsw+2121/list-1"

while nextURL != "terminate":
    data = fetch_and_parse(nextURL , headers, cookies)
    a_tags = data.find_all('a', class_='details-link residential-card__details-link')
    for a in a_tags:
        href = a.get('href')
        if href:
            href_list.append("https://www.realestate.com.au" + href)
            
    a_tags = data.find_all('a', class_='details-link child-listing__link')
    for a in a_tags:
        href = a.get('href')
        if href:
            href_list.append("https://www.realestate.com.au" + href)
   
    next = data.find_all("a",{"title": "Go to Next Page"})
    if not next :
        nextURL = "terminate"
    else:
        nextURL = "https://www.realestate.com.au" + next[0]['href']
```

```python
def extract_property_details(soup):
    property_details = {}

    # Extract the property address
    address_tag = soup.find('h1', class_='property-info-address')
    if address_tag:
        property_details['address'] = address_tag.text.strip()

    # Extract property type
    property_type_tag = soup.find('span', class_='property-info__property-type')
    if property_type_tag:
        property_details['property_type'] = property_type_tag.text.strip()
        
    # Extract property Detail
    detail_row_tag = soup.find('div', class_="Inline__InlineContainer-sc-lf7x8d-0 gOaIBl View__InlineStyled-sc-4or7us-0 AYXFA property-info__property-attributes")
    if detail_row_tag:
        values = detail_row_tag.find_all('p', class_="Text__Typography-sc-vzn7fr-0 OdxXk")
        
        property_details['bedrooms'] = values[0].text.strip().lower() if len(values) > 0 else None
        property_details['bathrooms'] = values[1].text.strip().lower() if len(values) > 1 else None
        property_details['parking_spaces'] = values[2].text.strip().lower() if len(values) > 2 else None

                
        sqrt_tag = detail_row_tag.find('span', class_="View__SpanWithEllipsisText-sc-1psmy31-2 iCVYXB")
        if sqrt_tag:
            property_details['square_footage'] = sqrt_tag.text.strip()
            
    # Extract additional property features
    feature_sections = soup.find('div', class_='property-features__section')
    if(feature_sections):
        for section in feature_sections:
            features = section.find_all('div', class_='property-features__feature')
            for feature in features:
                feature_text = feature.text.strip().lower()
                print(feature_text)
                if 'building size' in feature_text:
                    property_details['building_size'] = feature.find('span', class_='property-features__value').text.strip()
                elif 'land size' in feature_text:
                    property_details['land_size'] = feature.find('span', class_='property-features__value').text.strip()
                elif 'open car spaces' in feature_text:
                    property_details['open_car_spaces'] = feature.find('span', class_='property-features__value').text.strip()
    
    # Extract Auction Details
    auction_tag = soup.find('div', class_='auction')
    if auction_tag:
        property_details['auction'] = True
        auction_date_tag = auction_tag.find('span', class_='calendar-event__date')
        auction_time_tag = auction_tag.find('span', class_='calendar-event__time')
        if auction_date_tag and auction_time_tag:
            property_details['auction_date'] = auction_date_tag.text.strip()
            property_details['auction_time'] = auction_time_tag.text.strip()
        else:
            property_details['auction_date'] = 'Not available'
            property_details['auction_time'] = 'Not available'
    else:
        property_details['auction'] = False
        
    # Extract property price
    price_tag = soup.find('span', class_='property-price property-info__price')
    if price_tag:
        property_details['price'] = price_tag.text.strip()
                        
    # Extracting agent name
    agent_details={}
    agent_name_tag = soup.find('a', class_='agent-info__name')
    if agent_name_tag:
        agent_details['name'] = agent_name_tag.text
        
    # Extracting agent rating
    agent_rating_tag = soup.find('span', class_='styles__AvgRatingText-sc-1aagj3x-0')
    if agent_rating_tag:
        agent_details['rating'] = agent_rating_tag.text
    
    # Extracting agent reviews
    agent_reviews_tag = soup.find('span', class_='styles__ReviewsText-sc-1aagj3x-1')
    if agent_reviews_tag:
        agent_details['reviews'] = agent_reviews_tag.text

    # Extracting agent phone number
    agent_phone_tag = soup.find('div', class_='phone').find('a')
    if agent_phone_tag:
        phone_href = agent_phone_tag.get('href')
        if phone_href:
            agent_details['phone'] = phone_href.split(':')[1].strip()
        
    # Extracting agent address
    agent_address_tag = soup.find('div', class_='sidebar-traffic-driver__detail-info')
    if agent_address_tag:
        agent_details['address'] = agent_address_tag.text.strip()
            
    property_details["Agent Details"]=agent_details;
    
    return property_details
```

```python
import csv
import json

# List to store property details
property_details_list = []
for href in href_list:
    soup = fetch_and_parse(href, headers, cookies)
    
    property_details = extract_property_details(soup)

    property_details_list.append(property_details)


```

```python
with open('property_details.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = property_details_list[0].keys() if property_details_list else []
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(property_details_list)

with open('property_details.json', 'w', encoding='utf-8') as jsonfile:
    json.dump(property_details_list, jsonfile, ensure_ascii=False, indent=4)
```
